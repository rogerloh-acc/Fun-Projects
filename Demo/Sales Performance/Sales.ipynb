```python
import pandas as pd 
```

```python
df = pd.read_excel ("https://github.com/rogerloh-acc/Fun-Projects/blob/main/Demo/Sales%20Performance/SalesData.xlsx?raw=true")
df
```

```python
df.dropna (inplace = True)
df
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header

df
```

```python
df["SalesAmount"] = df["Quantity"] * df["UnitPrice"]
df
```

```python
df.info()
```

```python
df [["Quantity", "UnitPrice", "SalesAmount"]] = df[["Quantity", "UnitPrice", "SalesAmount"]].astype ("float")
df.info()
```

```python
salesperson = df.groupby ("Salesperson").sum()
salesperson.sort_values ("SalesAmount", ascending = False, inplace = True)
salesperson ["Sales %"] = salesperson ["SalesAmount"] / salesperson ["SalesAmount"].sum () * 100
salesperson ["Cum %"] = salesperson ["Sales %"].cumsum()
salesperson
```

```python
channel = df.groupby ("Channel").sum()
channel.sort_values ("SalesAmount", ascending = False, inplace = True)
channel ["Sales %"] = channel ["SalesAmount"] / channel ["SalesAmount"].sum () * 100
channel ["Cum %"] = channel ["Sales %"].cumsum()
channel
```
